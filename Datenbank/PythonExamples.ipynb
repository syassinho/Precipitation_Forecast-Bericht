## Funktion mit mehreren Rückgaben

```python
import math
def Kreismessung(radius):
    umfang=2*math.pi*radius
    flaeche=math.pi*radius**2
    return umfang,flaeche

U,A=Kreismessung(12)
print("Umfang=",U , "Fläche=", A)
print("Umfang=",U , "Fläche=",f"{A:8.2f}")#f-String literals
print(Kreismessung(25))
```

# Schleifen mit Fortschrittsanzeige: tqdm

```python
from tqdm import tqdm

n=10000
description="berechne Wurzeln..."
wurzeln=list()

if(n<1000):
    
    for i,x in enumerate(range(n)):
        wurzeln.append(x**0.5)
        print("berechne Wurzeln...", int(i/n*100), "%", end='\r')
    print("")
    print(wurzeln[4], wurzeln[144], wurzeln[n-1])


print ("tqdm....")
wurzeln=list()
for x in tqdm(range(n), desc=description):
    wurzeln.append(x**0.5)

print(wurzeln[4], wurzeln[144], wurzeln[n-1])
```

## PySerial

```python
#Kommunikation mit einem Thiesgerät....

import serial

ser = serial.Serial()

ser.baudrate = 9600
ser.port = 'COM5'
ser.stopbits = serial.STOPBITS_ONE
ser.bytesize = 8
ser.parity = serial.PARITY_NONE
ser.rtscts = 0 #hardware flow control off

ser.open()

ser.write("00TR4\r\n".encode())
res=ser.readline()
print(res)
ser.close()
```

## Dateien lesen und schreiben

```python
#this filters out lines with the wrong number of TAB-separated columns and stores to a new file.

fin = open("Palas_13415_2023-10-18.dat", "r")
fout = open("Palas_13415_2023-10-18_B.dat", "w")#warning! this is not append! use "a" instead of "w" to append. use "wb" etc. for binary file writing.

for line in fin:
    #split by tab
    fields=line.split('\t')
    #filter out lines with wrong number of fields
    if (len(fields)==107):
        #save to new file
        fout.write(line)
        fout.write("\n")

fin.close()
fout.close()

print("done.")
```
